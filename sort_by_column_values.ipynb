* https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html

* educational attainmentの降順にソート

```python
df.sort_values('Educational attainment', ascending=False)
```

* educational attainmentの大きい順に10カ国を求める

```python
df.sort_values('Educational attainment', ascending=False)[:10].index
```

* educational attainmentの値の上位10件の平均値

```python
df.sort_values('Educational attainment', ascending=False)[:10]['Educational attainment'].mean()
```
