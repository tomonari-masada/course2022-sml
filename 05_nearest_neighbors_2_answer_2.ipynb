# 訓練データにだけスケーラをfitさせる

```python
import numpy as np
import pandas as pd

%config InlineBackend.figure_format = 'retina'
```

* 日本をtest setとして除外し、残りのデータ集合を使う。
 * 日本の生活満足度を予測することを、最終的な目標としたのだった。

```python
df = pd.read_csv('lifesat_extended.csv', index_col=0)
df_test = df.loc[['Japan']]
df = df.drop(['Japan'])
```

* テストデータ以外の特徴量を`X`、ターゲットを`y`にセットする。

```python
X = df[['GDP per capita', 'Employment rate']]
y = df['Life satisfaction']
```

* 検証データとする一つの国を除いて、スケーラをfitさせる。

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_dist = pd.DataFrame(index=X.index)
for country in X.index:
  X_scaled = X.copy()
  scaler.fit(X.drop(country)) # ここでdropするのがポイント。
  X_scaled[X.columns] = scaler.transform(X)
  X_dist['dist ' + country] = np.linalg.norm(X_scaled - X_scaled.loc[country], axis=1)
```

```python
X_dist
```

* 以下のコードは、検証データを含めてスケーラをfitさせた場合のコードと、全く同じ。

```python
country = 'Korea'
sorted_indices = X_dist.sort_values('dist ' + country).drop(country).index

errors = []
for k in range(1, 28):
  error = abs(y[country] - y[sorted_indices[:k]].mean())
  errors.append(error)
errors = np.array(errors)
print(f'min error when k={(errors.argsort() + 1)[0]}')
```

```python
for country in X_dist.index:
  sorted_indices = X_dist.sort_values('dist ' + country).drop(country).index
  errors = []
  for k in range(1, 28):
    errors.append(abs(y[country] - y[sorted_indices[:k]].mean()))
  errors = np.array(errors)
  print(f'{country} k={(errors.argsort() + 1)[0]}')
```

```python
table = []
for country in X_dist.index:
  sorted_indices = X_dist.sort_values('dist ' + country).drop(country).index
  errors = []
  for k in range(1, 28):
    errors.append(abs(y[country] - y[sorted_indices[:k]].mean()))
  table.append(errors)
table = np.array(table)

df_errors = pd.DataFrame(data=table, index=X_dist.index, columns=range(1,28))
```

```python
df_errors
```

```python
df_errors.mean(axis=0)
```

```python
k = df_errors.mean(axis=0).idxmin()
print(k)
```

```python
X_test = df_test[['GDP per capita', 'Employment rate']]

X_scaled = X.copy()
X_scaled[X.columns] = scaler.fit_transform(X)

X_test_scaled = X_test.copy()
X_test_scaled[X_test.columns] = scaler.transform(X_test)
```

```python
X_scaled['distances'] = np.linalg.norm(X_scaled[X_test_scaled.columns] - X_test_scaled.loc['Japan'], axis=1)
```

```python
sorted_indices = X_scaled.sort_values('distances').index
prediction = y[sorted_indices[:k]].mean()
error = abs(df_test['Life satisfaction'] - prediction)
print(f'k={k} ; error={error.values[0]:.3f}')
```
